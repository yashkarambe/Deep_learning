```python
import tensorflow as tf
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense , Flatten
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
```

```python
(x_train ,y_train ) , (x_test , y_test) = keras.datasets.mnist.load_data()
```

```python
x_test.shape
```

```python
x_train = x_train/255
x_test = x_test/255
x_train.shape
```

```python
y_train.shape
```

```python
plt.imshow(x_train[1])
```

```python
model = Sequential()
model.add(Flatten(input_shape=( 28, 28)))
model.add(Dense(128 , activation='relu'))
model.add(Dense(32 , activation='relu'))
model.add(Dense(10 , activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(loss=keras.losses.sparse_categorical_crossentropy,optimizer=keras.optimizers.legacy.Adam(),metrics=['accuracy'])
```

* model.compile(loss='sparse_categorical_crossentropy' , optimizer=keras.optimizers.legacy.Adam() , metrics=['accuracy'])

```python
history = model.fit( x_train , y_train , epochs=10 , validation_split=0.2 )
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred = y_pred.argmax(axis=1)
```

```python
accuracy_score(y_test , y_pred)
```

```python
plt.plot(history.history['val_loss'] )
plt.plot(history.history['loss'])
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
```
